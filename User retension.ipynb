# Cohort Analysis (single dataframe)

## Create cohort months

```python
# create a new column called TransactionMonth, get month value from each transaction and put the first day of the month to the TransactionMonth
def get_month(x):
    return dt.datetime(x.year, x.month, 1)
trans['TransactionMonth']=trans['transaction_date'].apply(get_month)
# there might be multiple transactions by each customer, we find out the month of first transaction for each customer
trans.groupby('customer_id')['TransactionMonth'].min()
trans.groupby('customer_id')['TransactionMonth'].transform('min') # we actually need to retain the orginal dataframe structure, so a transform() is used
```

```python
# create CohortMonth: cohort is devided and measured by months
trans['CohortMonth']= trans.groupby('customer_id')['TransactionMonth'].transform('min')
print(trans.head())
```

## create CohortIndex: 

```python
def get_date(df, column):
    year = df[column].dt.year
    month = df[column].dt.month
    day = df[column].dt.day
    return year, month, day
trans_year, trans_month, _ = get_date(trans, 'TransactionMonth') # here if we need to discard the day, we need to place a placeholder to avoid value mismatches 
cohort_year, cohort_month, _ = get_date(trans,'CohortMonth')
```

```python
# creating cohort index
years_diff = trans_year - cohort_year
months_diff = trans_month - cohort_month
trans['CohortIndex'] = years_diff*12 + months_diff + 1
trans.head()
```

```python
cohort_data = trans.groupby(['CohortMonth', 'CohortIndex']).nunique()['customer_id'].reset_index() #calculate the no. of unique customers in each CohortMonths and CohortIndex (aka unique customer count)
cohort_data.head(5)
```

## calculate the retention rate and create visualisation

```python
cohort_counts = cohort_data.pivot(index='CohortMonth', columns = 'CohortIndex', values= 'customer_id') #make a pivot table that has the customer count sorted according to the CohortMonth and CohortIndex
cohort_sizes = cohort_counts.iloc[:,0] # take from the column index 1 (the original no. of customers) as the cohort size
retention = cohort_counts.divide(cohort_sizes, axis=0).round(3)*100 # axis=0 is needed as column_sizes is a column index (so by default divide() will align with this), otherwise pandas will divide along columns
retention
```

```python
print(retention.index)
```

```python
retention.index = retention.index.strftime('%Y-%m') # converts the CohortMonth into a Year-month format, getting rid of the date
plt.figure(figsize=(16,10))
plt.title('Average Standard Cost: Monthly Cohorts', fontsize = 14)
sns.heatmap(retention, annot = True, vmin = 0.0, vmax = 100, cmap = 'YlGnBu', fmt = 'g')
plt.ylabel('Cohort Month')
plt.xlabel('Cohort Index')
plt.yticks(rotation = 360)
plt.savefig('cohort_analysis.png')
plt.show()
```

# RFM Analysis (merged)

## Calculate RFM Scores

```python
# Monetary: calculate the total profit of each customer and rank them from high profit to low profit (contributed)
cust_info['profit']= cust_info['list_price'] - cust_info['standard_cost']
mon = cust_info.groupby(['customer_id'])['profit'].sum().reset_index()
mon['Rank_M'] = mon['profit'].rank(ascending = False)
mon.head()
```

```python
# Recency: 
rec = cust_info.groupby(['customer_id'])['transaction_date'].max().reset_index()
rec['Rank_R']= rec['transaction_date'].rank(ascending=False)
rec.head()
```

```python
# Frequency
fre = cust_info.groupby(['customer_id'])['transaction_date'].count().reset_index()
fre['Rank_F']= fre['transaction_date'].rank(ascending = False)
fre.head()
```

```python
df_rank= mon.merge(rec, how ='left', on= 'customer_id').merge(fre, how = 'left', on ='customer_id')
df_rank['RFM_Score']= df_rank['Rank_M']+df_rank['Rank_R']+df_rank['Rank_F'] 
df_rank['RFM_Score']=df_rank['RFM_Score'].rank(ascending = False)
df_final= cust_info.merge(df_rank, how = 'left', on='customer_id')


df_final['NetValue'] = df_final['RFM_Score'].apply(lambda x: 'High' if x <= 1000 else 'Medium' if 1000 < x <= 3000 else 'Low' if 3000 < x <= 4000 else 'Very Low')
df_final['NetValue']
print(df_final['NetValue'])
```
